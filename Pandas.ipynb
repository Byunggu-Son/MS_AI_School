```python
# Group By를 이용한 계산 및 요약 통계
german_sample = german[['Credit Amount','Type of apartment']]
german_sample
```

```python
german_grouped = german_sample['Credit Amount'].groupby(german_sample['Type of apartment'])
```

```python
german_grouped.mean()
```

```python
# groupby한건 뒤에 mean과 max등과 같이 나와야한다!
german_grouped.describe()
```

```python
german_grouped.max()
```

```python
german_sample = german[['Type of apartment','Sex & Marital Status','Credit Amount']]
```

```python
for type, group in german_sample.groupby('Type of apartment'):
  print(type) # 얘만 출력하면 1,2,3만 나옴(파이썬의 특징)
  print(group.head())
```

```python
for type, sex, group in german_sample.groupby(['Type of apartment','Sex & Marital Status']):
  print(type,sex) # 넘어 오는건 2개인데 for뒤에는 3개를 썼기에 오류!
```

```python
for (type, sex), group in german_sample.groupby(['Type of apartment','Sex & Marital Status']):
  print(type,sex) # 얘만 하면 그냥 1 2 3 4 정도만 나옴. 즉 앞에꺼 혹은 ()로 묶은 애들은 튜플로 넘어오고, 맨 뒤에서 데이터 값이 온다. 즉,(1,1) (1,2)로 넘어오면서 각각의 데이터는 여기서는 group을 통해서 온다.
  print(group.head(n=3))
```

```python
births = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv')
```

```python
births.shape
```

```python
births.head()
```

```python
# 10년 단위의 새로운 컬럼 추가
births['decade'] = births['year']//10 * 10
```

```python
births.head()
```

```python
births.pivot_table('births', index='decade', columns='gender', aggfunc='sum') # pivot 모니터 생각 , agg는 집계인 aggregation.
#pivot_table(집계대상,인덱스,컬럼, 집계방식)
```

```python
import matplotlib.pyplot as plt
```
