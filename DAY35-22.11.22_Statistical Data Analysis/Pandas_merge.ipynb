```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    '날짜' : [1101, 1102, 1103, 1104, 1105],
    '온도' : [20, 18, 16, 14, 5]
}, columns=['날짜','온도'])
```

```python
df1
```

```python
df2 = pd.DataFrame(data={
    '날짜' : [1101, 1102, 1103, 1104, 1109, 1120],
    '습도' : [10, 20, 40, 100, 0, 50]
}, columns=['날짜','습도'])
```

```python
df2
```

```python
df3 = df1 + df2
df3
```

## merge 

```python
#디폴트값으로 병합시 중복되는 부분만 병합됨.
pd.merge(df1,df2)
```

```python
# 겹치지 않는 부분도 합치고 싶다면 how=outer
pd.merge(df1,df2, how='outer')
```

```python
pd.merge(df1, df2, how='left') # 맨 왼쪽 데이터에 기준을 맞춰서 병합
```

```python
pd.merge(df1, df2, how='right') #맨 오른쪽 데이터에 기준을 맞춰서 병합
```

```python
df3 = pd.DataFrame({
    'date' : [1101, 1102, 1103, 1104, 1105],
    '온도' : [20, 18, 16, 14, 5]
}, columns=['date','온도'])
```

```python
df4 = pd.DataFrame({
    '날짜' : [1101, 1102, 1103, 1104, 1105],
    '온도' : [20, 18, 16, 14, 5]
}, columns=['날짜','온도'])
```

```python
pd.merge(df3, df4, left_on='date', right_on='날짜')
```
