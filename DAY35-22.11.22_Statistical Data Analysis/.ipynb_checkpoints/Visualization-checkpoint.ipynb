## 바 그래프

```python
data = {
    '바나나' : 3, '사과' : 2, '오렌지' : 10
}
# 선: plot, 바:bar
print(data.keys(), data.values())
plt.bar(data.keys(), data.values())
plt.title('과일 개수')
plt.xticks([0,1,2], data.keys()) # x에 해당하는 라벨(순서 변경도 가능)
plt.yticks(sorted(data.values()))
plt.grid(True)
plt.xlabel('과일')
plt.ylabel('개수')
plt.show()
```

```python
# 세로막대그래프 barh, x값&y값 변경
data = {
    '바나나' : 3, '사과' : 2, '오렌지' : 10
}
# 선: plot, 바:bar
print(data.keys(), data.values())
plt.barh([0,1,2], data.values())
plt.title('과일 개수')
plt.yticks([0,1,2], data.keys()) # x에 해당하는 라벨(순서 변경도 가능)
plt.xticks(sorted(data.values()))
plt.grid(True)
plt.xlabel('과일')
plt.ylabel('개수')
plt.show()
# 바그래프는 이거 정도면 끝.
# barh를 쓰면 데이터 순서만 살짝 수정.
```

## 파이차트

```python
ratio = [34, 32, 16, 18] # 비율
labels = ['Apple', 'Banana', 'Melon', 'Grapes'] # 라벨 이름
explode = [0.05, 0.05, 0.05, 0.05] # 중심에서 벗어나는 비율
colors = ['silver', 'gold', 'whitesmoke', 'lightgray'] # 색상
# 파이차트 메서드는 plt.pie

# startangle : 시작 지점(0이면 양의 x축부터 시작)
# counterclock : True(시계방향), False(시계반대방향)
# shadow : 그림자 여부

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)

plt.show()
```

```python
# 부채꼴로 그리기 위해서는 wedgeprops 옵션
wedgeprops={'width': 0.7, 'edgecolor': 'w'}
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 예제
# 정규분포 생성(중앙값이 2, 표준편차2인 랜덤 데이터 100개 생성)
data_a = np.random.normal(2.0, 2.0, 100)
```

```python
estimator = stats.gaussian_kde(data_a, bw_method='silverman') # 가우시안 커널 밀도 추정
X = np.arange(-10,10)
K = estimator(X)
```

```python
plt.plot(X, K, label='gaussian_kde')
plt.legend(loc='best')
plt.show()
```

# Scatter plot
#### 다변량 데이터의 수치적 변수들 간의 관계 확인 가능
##### 각 값에 대해 순서쌍으로 표기하여 2차원 평면상에 점으로 표기


```python
# seed는 난수 생성 시 키값을 통해 난수 고정(seed를 안써주면 랜덤시 다른값이 뜨지만 seed를 쓰면 키값으로 난수를 고정시킴.)
np.random.seed(0)
# 50개 난수 생성
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
```

```python
# # seed()안썼을 경우
# np.random.rand(1)
# np.random.rand(2)
# np.random.rand(3)

# # seed()썼을 경우
# np.random.seed(1)
# np.random.rand(1)
# np.random.rand(2)
# np.random.rand(3)
```

```python
# 점 사이즈 랜덤 값
area = (30 * np.random.rand(n))**2
# 색상 랜덤 값
colors = np.random.rand(n)


'''
cmap : 컬러맵 지정 c로 넘겨받은 값을 토대로 자동으로 색상 지정
cmap 종류 : Spectral, rainbow, autumn ...
c : colors 색
s : size에 따른 원의 크기
'''
plt.scatter(x, y, c=colors, s=area, alpha=0.5, cmap='Spectral')
# colorbar 생성
plt.colorbar()
plt.show()
```

# 박스 그래프

```python
# 1. 기본 스타일 설정
plt.style.use('default')

# plt.rcParams['figure.figsize'] = (4, 3)
# plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
'''
np.random.normal -> 정규분포 데이터 생성
mu : 중앙값
sigma : 표준편차
size : 결과 사이즈
'''
# 중앙값이 0이고 표준편차가 2인 랜덤 데이터 1000개 생성
data_a = np.random.normal(0, 2.0, 1000)
# 중앙값이 -3이고 표준편차가 1.5인 랜덤 데이터 500개 생성
data_b = np.random.normal(-3.0, 1.5, 500)
# 중앙값이 1.2이고 표준편차가 1.5인 랜덤 데이터 1500개 생성
data_c = np.random.normal(1.2, 1.5, 1500)
# 3. 그래프 그리기
fig, ax = plt.subplots()
'''
notch : bool (중앙값의 95%신뢰구간을 노치형태로 나타냄)
whis : float (수염의 길이를 정할 때 IQR에 곱해지는 변수 default : 1.5)
vert : bool (False(가로) 혹은 True(세로))
'''

# 박스 그래프 : boxplot
ax.boxplot([data_a, data_b, data_c], notch=True, whis=1.5, vert=True) # vert=False 가로 박스
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
plt.show()
```

```python
# 가로 박스 그래프
fig, ax = plt.subplots()
ax.boxplot([data_a, data_b, data_c], notch=True, whis=1.5, vert=False)
ax.set_xlim(-10.0, 10.0)
# ax.set_xlabel('Data Type')
# ax.set_ylabel('Value')
plt.show()
```

# 바이올린 플롯(Violin plot)

#### 박스플롯과 동일하게 일변량, 연속형 데이터 분포를 설명하기 위한 그래프
#### 대부분의 내용이 박스 플롯과 동일
#### 범주에 따라 다른 분포의 실제 데이터 또는 전체 형상을 보여줌
#### 박스 플롯보다 더 실제에 가까운 분포를 알 수 있음

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 기본 스타일 설정
plt.style.use('default')
# plt.rcParams['figure.figsize'] = (4, 3)
# plt.rcParams['font.size'] = 12

# 2. 데이터 준비
np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

# 3. 그래프 그리기
'''
showmeans : bool (중앙값 위치 직선 표시 여부)
showextrema : bool (최대 최소값 직선 표시 여부)
showmedians : bool (중간값 직선 표시 여부)
quantiles : list (데이터 분포 분위수 표시 여부) ex. quantiles=[[0,3,0,7],[0.2, 0.8], [0.1,0.9]]
'''
violin = plt.violinplot([data_a, data_b, data_c], showmeans=True, showmedians=True, quantiles=[[0.3,0.7],[0.2, 0.8], [0.1,0.9]])
plt.ylim(-10.0, 10.0)
plt.xticks(np.arange(1, 4), labels=['A','B','C'])
plt.xlabel('Data Type')
plt.ylabel('Value')
# 색상 지정
'''
bodies: 각각의 바이올린 분포의 채워진 영역을 포함하는 PolyCollection 인스턴스의 리스트.
cmeans: 바이올린 분포의 평균값을 표시하는 LineCollection 인스턴스.
cmins: 바이올린 분포의 최소값을 표시하는 LineCollection 인스턴스.
cmaxes: 바비올린 분포의 최대값을 표시하는 LineCollection 인스턴스.
cbars: 바이올린 분포의 중심을 표시하는 LineCollection 인스턴스.
cmedians: 바이올린 분포의 중간값을 표시하는 LineCollection 인스턴스.
cquantiles: 바이올린 분포의 분위값을 표시하는 LineCollection 인스턴스
'''
violin['bodies'][0].set_facecolor('blue')
violin['bodies'][1].set_facecolor('red')
violin['bodies'][2].set_facecolor('green')
violin['cbars'].set_edgecolor('red')
violin['cmaxes'].set_edgecolor('orange')
violin['cmins'].set_edgecolor('orange')
violin['cmeans'].set_edgecolor('white')
violin['cmedians'].set_edgecolor('yellow')
violin['cquantiles'].set_edgecolor('black')
plt.show()
```
