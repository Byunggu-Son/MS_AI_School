```python
plt.plot(x, np.sin(x-0), color='blue') #x라는 배열 전체의 값에 -0을 한 것을 의미
plt.plot(x, np.sin(x-1), color='g')  #색상 지정 방법 : 이름
plt.plot(x, np.sin(x-2), color='k') # 약어 black
plt.plot(x, np.sin(x-3), color='0.75') # 0 ~ 1사이의 회색조를 표현한다.
plt.plot(x, np.sin(x-4), color='#FFDD44') # 16진수로 표현한 RGB
plt.plot(x, np.sin(x-5), color=(1.0,0.2,0.3)) # RGB값을 튜플로 표현
plt.plot(x, np.sin(x-2), color='chartreuse') # 이름으로..
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-') #기호로 표시
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x,x+0,'-g') # 한번에 통합해서 solid라인에 green이라는 뜻. 같이 사용하려면 둘다 약어로 해야됨.
plt.plot(x,x+1,'--c')
plt.plot(x,x+2,'-.k')
plt.plot(x,x+3,':r')
```

```python
# 플롯에 레이블 달기
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x')
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)') # 앞에 x,y안붙이고 쓰면 선을 알려주는 label legend와 함께 써야됨.
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()
```

```python
# 산점도의 출력

x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y,'o', color='black')
```

```python
rng = np.random.RandomState(0) #안에 들어가있는 seed값에 따라서 값이 달라짐. 같은 값을 여러번 돌려도 결과는 똑같음.
for marker in ['o',',','x','+','v','<','>','s','d']:
  plt.plot(rng.rand(5), rng.rand(5),marker,label='marker={0}'.format(marker))
  plt.legend()
  plt.xlim(0,1.8) # 범례 자리를 침범하지않게 조정했음.
```

```python
plt.plot(x,y,'-ok') # -ok : solid에 산점도, 색상은 검은색
```

```python
plt.plot(x,y, '-p', color='gray',#p는 펜타곤
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='red',
         markeredgewidth='2') 
```

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
color = rng.randn(100)
size = 1000 * rng.randn(100)

plt.scatter(x,y,s=size,alpha=0.3, c=color, cmap='viridis',) 
# 사이즈 지정 s, 색상 지정 c, 투명도 지정 : 알파값 지정, 색깔 다채롭게는 cmap검색 참조(앞쪽 컬러에 부여한 숫자에 따라 색깔 표현하는 듯.)
plt.colorbar() # 우측 컬러바 생성
```

```python
from sklearn.datasets import load_iris # datasets 안에는 유명한 데이터셋이 들어가있어 파이썬에서 제공한다.

iris = load_iris()
```

```python
# iris.shape 이 안되면 우리가 아는 numpy배열이 아니구나라고 생각
type(iris) # 우리가 아는 numpy배열이 아니구나라고 생각
```

```python
# iris.data 
# iris.data.T 전체 배열 봄
#iris.target
features = iris.data.T
```

```python
plt.scatter(features[0],features[1],s=100 * features[3], c=iris.target, alpha=0.3, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
# 오차 시각화
x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) * np.random.randn(50)

plt.errorbar(x,y, yerr=dy, fmt='.k') # fmt는 선이나 마커의 포맷
```

```python
plt.errorbar(x,y,yerr=dy, fmt='o', color='black', ecolor='lightgray', elinewidth=3, capsize=0) #ecolor 오차범위 색상, capsize:위아래 양 끝 캡 사이즈 조절
```
