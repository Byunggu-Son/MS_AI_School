```python
import bs4
import requests
```

```python
url = 'https://search.shopping.naver.com/search/category/100005307'
result = requests.get(url).text
result
```

```python
bsObj = bs4.BeautifulSoup(result, 'html.parser') #bs4는 여러가지 parsing하는데 그 중 하나가 html (그 외 xml)
```

```python
items = bsObj.find_all('a',{'class':'basicList_link__JLQJf'}) # find하면 해당하는 그거 하나. find_all은 전체
```

```python
for item in items:
  print(item.text)
```

```python
# 위에서 정상적으로 나오는 값은 4개이므로 4개만 보도록 하자
print(items[0].text)
print(items[1].text)
print(items[2].text)
print(items[3].text)
```
